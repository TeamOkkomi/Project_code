```python
Ndf_2011 = df_2011
Ndf_2012 = df_2012
Ndf_2013 = df_2013
Ndf_2014 = df_2014
Ndf_2015 = df_2015
Ndf_2016 = df_2016
Ndf_2017 = df_2017
Ndf_2018 = df_2018
Ndf_2019 = df_2019
Ndf_2020 = df_2020
Ndf_2021 = df_2021
Ndf_2022 = df_2022
```

```python
# 2011년~2022년까지를 전부 붙이면     # 레퍼 https://clolee.tistory.com/14
Ndf_list_for_concat = [Ndf_2011, Ndf_2012, Ndf_2013, Ndf_2014, Ndf_2015, Ndf_2016, Ndf_2017, Ndf_2018, Ndf_2019, Ndf_2020, Ndf_2021, Ndf_2022]
Ndf_All_concatted = pd.concat(Ndf_list_for_concat, ignore_index=True)
Ndf_All_concatted
# 1131줄. 이제 여기서 특보강도=0을 전부 빼고, 인덱스 새로 만들어주고, 인덱스가 홀수인 행은 전부 빼야함
```

```python
# 특보강도 0인 행 전부 삭제하기 dfresult = df명[ ~df명['컬럼명'].str.contains("제거대상문자열", na=False, case=False) ]


#  특정문자 마스크 씌워서 안보이게 하기. 레퍼 https://codechacha.com/ko/python-convert-integer-to-string/
# mask에다가 어떤 칼럼명에 어떤 값을 갖고 있는 애들을 선택하게 한다.
# mask = df['컬럼명'].isin(['특정값'])

# 그리고 마스크 영역을 제외한 영역을 볼 수 있다.
# df[~mask].head() 

# 저장하려면 객체에 다시 저장
# df1 = df[~mask] 


mask_all = Ndf_All_concatted['특보강도'].isin([0])
Ndf_All_concatted_masked = Ndf_All_concatted[~mask_all]
```

```python
#  인덱스 리셋. 레퍼 https://kongdols-room.tistory.com/123 <=아근데 왜 드래그를 막아놧어 진짜짜증
# df명.reset_index(drop=True, inplace=False)
Ndf_All_concatted_masked = Ndf_All_concatted_masked.reset_index(drop=True, inplace=False)
Ndf_All_concatted_masked
```

```python
#  인덱스가 홀수인 행 삭제. 레퍼 https://dojang.io/mod/page/view.php?id=2211  <=야 이건 튜플 예시인데 해주네
# df명[1:len(df명):2]
Ndf_All_concatted_masked_withoutOddNum = Ndf_All_concatted_masked[1:len(Ndf_All_concatted_masked):2]
Ndf_All_concatted_masked_withoutOddNum
```

```python
Ndf_All_concatted_masked_withoutOddNum_newIDX = Ndf_All_concatted_masked_withoutOddNum.reset_index(drop=True, inplace=False)
Ndf_All_concatted_masked_withoutOddNum_newIDX
```

```python
# 이름이 너무 길다. mdf로 이름 다시 짓는다
mdf = Ndf_All_concatted_masked_withoutOddNum_newIDX
mdf
```

## 자 이제부터는 특보강도를 전부 주의보 경보 문자열로 바꾼다
> * 이게모하는짓인지...

```python
mdf = mdf.astype('str')
mdf.info()
```

```python
mdf['특보강도'] = mdf['특보강도'].replace('1', '풍랑주의보')
mdf['특보강도'] = mdf['특보강도'].replace('2', '풍랑경보')
mdf['특보강도'] = mdf['특보강도'].replace('3', '태풍주의보')
mdf['특보강도'] = mdf['특보강도'].replace('4', '태풍경보')
mdf
```
