```python
import pandas as pd

df1 = pd.read_csv('./csv/모델1_데이터셋/2011.csv',  encoding='utf-8', index_col='일시')
df1
```

```python
df2 = pd.read_csv('./csv/모델1_데이터셋/2012.csv',  encoding='utf-8', index_col='일시')
df3 = pd.read_csv('./csv/모델1_데이터셋/2013.csv',  encoding='utf-8', index_col='일시')
df4 = pd.read_csv('./csv/모델1_데이터셋/2014.csv',  encoding='utf-8', index_col='일시')
df5 = pd.read_csv('./csv/모델1_데이터셋/2015.csv',  encoding='utf-8', index_col='일시')
df6 = pd.read_csv('./csv/모델1_데이터셋/2016.csv',  encoding='utf-8', index_col='일시')
df7 = pd.read_csv('./csv/모델1_데이터셋/2017.csv',  encoding='utf-8', index_col='일시')
df8 = pd.read_csv('./csv/모델1_데이터셋/2018.csv',  encoding='utf-8', index_col='일시')
```

```python
import pandas as pd

df = pd.read_csv('./csv/모델1_데이터셋/2011~2020_m1.csv', encoding='utf-8')
df
```

```python
df = df.drop(['특보강도'], axis=1)
df
```

```python
df_timeleft1 = pd.read_csv('./csv/모델2_데이터셋/2011.csv', encoding='utf-8')
df_timeleft2 = pd.read_csv('./csv/모델2_데이터셋/2012.csv', encoding='utf-8')
df_timeleft3 = pd.read_csv('./csv/모델2_데이터셋/2013.csv', encoding='utf-8')
df_timeleft4 = pd.read_csv('./csv/모델2_데이터셋/2014.csv', encoding='utf-8')
df_timeleft5 = pd.read_csv('./csv/모델2_데이터셋/2015.csv', encoding='utf-8')
df_timeleft6 = pd.read_csv('./csv/모델2_데이터셋/2016.csv', encoding='utf-8')
df_timeleft7 = pd.read_csv('./csv/모델2_데이터셋/2017.csv', encoding='utf-8')
df_timeleft8 = pd.read_csv('./csv/모델2_데이터셋/2018.csv', encoding='utf-8')
```

```python
df_timeleft8
```

```python
#데이터셋 합치기
df_timeleft = pd.concat([df_timeleft1, df_timeleft2, df_timeleft3, df_timeleft4, df_timeleft5, df_timeleft6, df_timeleft7, df_timeleft8])
df_timeleft
```

```python
df_timeleft = df_timeleft[['일시','특보강도','남은시간']]
df_timeleft
```

```python
# df + 남은시간, 특보강도
df = df.merge(df_timeleft, on='일시', how='left')
df
```

```python
df.tail(30)
```

```python
#일시를 인덱스로 설정
df = df.set_index('일시')
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
#결측치 0으로 대체
df = df.fillna(value={'남은시간':0, '특보강도':0})
df
```

```python
df.isna().sum()
```

```python
#모델2 데이터셋(특보강도0~4) csv 저장
df.to_csv('./csv/모델2_데이터셋/2011~2020(특보강도0~4).csv')
```

```python
#모델2 데이터셋(특보강도 0, 1, 2)
df_m2 = df.replace(to_replace={'특보강도':{3:1, 4:2}})
df_m2
```

```python
df_m2.info()
```

```python
df_m2.describe()
```

```python
#csv 저장
df_m2.to_csv('./csv/모델2_데이터셋/2011~2020_m2.csv')
```
